```python
#biblioteca usada
import pandas as pd
import requests
import zipfile
import io
```

```python
#Extração das informações da CVM

r = requests.get("http://dados.cvm.gov.br/dados/CIA_ESTRANG/CAD/DADOS/cad_cia_estrang.csv")
r.text.split("\n") 
```

```python
#Organização dos dados extraídos
[i.strip (). split(";") for i  in r.text.split("\n")  ]
```

```python
#Armazenando o resultado anterior em uma variável "Linhas"
lines = [i.strip (). split(";") for i  in r.text.split("\n")  ]
```

```python
#Criando o DF
df = pd.DataFrame(lines[1:], columns= lines[0])
```
